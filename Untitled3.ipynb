```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, dct
import scipy.fft
```

```python
N = 50
jump = np.zeros(N)
jump[0] = 1

single_impulse = np.zeros(N) 
for i in range(N//2,N):
    single_impulse[i] = 1
```

```python
def KiX(x, a):
    yn = x[1:]* a[0] + x[: -1] * a[1]
    y = np.concatenate((x[0:1] * a[0], yn))
    return y
```

```python
def BiX(x, a):
    y1 = 0
    y = np.zeros_like(x, dtype=float)
    for i in range(len(x)):
        y[i] = x[i] + y1 * a
        y1 = y[i]
    return y
```

```python
a = [0.7,0.4]
impuls_KiX = KiX(single_impulse,a)
plt.figure(figsize=(16, 4), dpi=120)
plt.title('KiX Импулс single_impulse ', fontsize=14)
plt.stem(impuls_KiX, use_line_collection=True, basefmt='C0')
plt.xlim([-0.5, N-0.5])
plt.grid(True)
```

```python
a = 0.7
impuls_BiX = BiX(single_impulse,a)
plt.figure(figsize=(16, 4), dpi=120)
plt.title('BiX Импулс single_impulse ', fontsize=14)
plt.stem(impuls_BiX, use_line_collection=True, basefmt='C0')
plt.xlim([-0.5, N-0.5])
plt.grid(True)
```

```python
B_fft = scipy.fft.fft(impuls_BiX)[:len(impuls_BiX) // 2]
amp = np.abs(B_fft)
ph = np.arctan2(B_fft.imag, B_fft.real)
plt.plot(amp)
# plt.plot(ph)
plt.show
```
